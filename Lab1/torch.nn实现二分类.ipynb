```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
import random
# 创建数据集
n_data = torch.ones(50, 2)
x1 = torch.normal(2 * n_data, 1)
y1 = torch.zeros(50)
x2 = torch.normal(-2 * n_data, 1)
y2 = torch.ones(50)
x = torch.cat((x1, x2), 0).type(torch.FloatTensor)
y = torch.cat((y1, y2), 0).type(torch.LongTensor)
# 定义一个简单的Sigmoid二分类模型
class SimpleSigmoidClassifier:
    def __init__(self, input_size):
        self.weights = torch.randn(input_size, requires_grad=True)
        self.bias = torch.randn(1, requires_grad=True)

    def forward(self, x):
        z = torch.dot(x, self.weights) + self.bias
        return torch.sigmoid(z)

# 准备数据
input_size = 10  # 输入特征的维度
data_size = 1000
X = torch.randn(data_size, input_size)
y = torch.randint(0, 2, (data_size, 1)).float()

# 初始化模型
model = SimpleSigmoidClassifier(input_size)

# 定义损失函数和优化器
criterion = nn.BCELoss()
optimizer = optim.SGD([model.weights, model.bias], lr=0.01)

# 训练模型
num_epochs = 100
for epoch in range(num_epochs):
    outputs = model.forward(X)
    loss = criterion(outputs, y)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if (epoch+1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item()}")

# 使用训练好的模型进行预测
with torch.no_grad():
    test_input = torch.randn(5, input_size)
    predictions = model.forward(test_input)
    binary_predictions = (predictions >= 0.5).float()
    print("Predictions:", binary_predictions)
```
